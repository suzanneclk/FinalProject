```python
import pandas as pd
import yfinance as yf
```

```python
stockData = pd.read_csv("constituents-financials_csv.csv")
stockData
```

```python
stockData["new data"] = 0
```

```python
newInfo = {}

""" 
I realized that there are some stock that do not have a balnce sheet I guess for some reason. So I implemented some error handling into our code to handle that - JB

"""

for c, r  in stockData.iterrows():
    companyBalanceSheet = yf.Ticker(r[0]).balance_sheet
    
    # print(companyBalanceSheet.index)
    try:
        currentAssets = companyBalanceSheet.loc["Current Assets"][0] #[1] is for 2022
    
    
        newInfo[r.Symbol] = currentAssets
         
    except:
        print(r.Symbol)
        continue
    
```

```python
for key, currentAssets in newInfo:
    print(key)
```
