```python
import pandas as pd
import yfinance as yf
```

```python
stockData = pd.read_csv("constituents-financials_csv.csv")
stockData
```

```python
stockData["new data"] = 0
```

```python
for c, r  in stockData.iterrows():
    companyBalanceSheet = yf.Ticker(r[0]).balance_sheet
    
    totalDebt = companyBalanceSheet.loc["Total Debt"][1] #[1] is for 2022
    
    # print(r.shape)
    
    stockData.iloc[c]["new data"] = totalDebt
    
    break

    
    
```
