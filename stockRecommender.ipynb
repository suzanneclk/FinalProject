I realized when trying to pull financals that there are some companies in the dataframe that are no longer on the sp500 due to being pulled or aquired so I went through and figured out which ones so that I can remove them and reduce the processing time

```python

symbolsNotFound = ['ATVI', 'AET', 'AFL', 'ALXN', 'AGN', 'ADS', 'ALL', 'AXP', 'AIG', 'AMP', 'ABC', 'APC', 'ANDV', 'ANTM', 'ARNC', 'AIZ', 'BHGE', 'BLL', 'BAC', 'BBT', 'BRK.B', 'BHF', 'BF.B', 'CA', 'COG', 'COF', 'CBG', 'CBS', 'CELG', 'CTL', 'CERN', 'SCHW', 'CB', 'CI', 'XEC', 'CINF', 'C', 'CFG', 'CTXS', 'CMA', 'CXO', 'CSRA', 'DFS', 'DISCA', 'DISCK', 'DISH', 'DWDP', 'DPS', 'DRE', 'ETFC', 'EVHC', 'RE', 'ESRX', 'FB', 'FITB', 'FISV', 'FLIR', 'FBHS', 'GGP', 'GS', 'HRS', 'HIG', 'HUM', 'HBAN', 'INFO', 'JEC', 'JPM', 'KSU', 'KEY', 'LB', 'LLL', 'LUK', 'LNC', 'L', 'MTB', 'MET', 'KORS', 'MON', 'MS', 'MYL', 'NAVI', 'NFX', 'NLSN', 'NBL', 'NTRS', 'PBCT', 'PKI', 'PNC', 'PCLN', 'PFG', 'PGR', 'PRU', 'RJF', 'RTN', 'RHT', 'RF', 'COL', 'SCG', 'SNI', 'STT', 'SYMC', 'SYF', 'BK', 'TRV', 'TIF', 'TWX', 'TMK', 'TSS', 'USB', 'UTX', 'UNM', 'VAR', 'VIAB', 'WFC', 'HCN', 'WLTW', 'WYN', 'XLNX', 'XL', 'ZION']
indeciesToDrop = [sp500[sp500["Symbol"].isin(symbolsNotFound)].index]

for index in indeciesToDrop:
    sp500 = sp500.drop(index=index)
```

```python
newInfo = {}
symbolsNotFoundInLoop = []

""" 
I realized that there are some stock that do not have a balnce sheet I guess for some reason. So I implemented some error handling into our code to handle that - JB

"""

for c, r  in sp500.iterrows():
    companyBalanceSheet = yf.Ticker(r[0]).balance_sheet
    
    # print(companyBalanceSheet.index)
    try:
        currentAssets = companyBalanceSheet.loc["Current Assets"][0] #[1] is for 2022
    
    
        newInfo[r.Symbol] = currentAssets
         
    except:
        symbolsNotFoundInLoop.append(r[0])

print(len(symbolsNotFoundInLoop)) #Shold be 0
    
```

```python
#Was using this to figure out which symbols we need to remove
print(symbolsNotFound)
#This is a good demonstration that there are some companies in this dataset that are no longer listed becuase of recent events
stockData[stockData['Symbol'] == 'ATVI']
```

```python
#Changing the symbol into a index
sp500 = sp500.set_index('Symbol')
```

```python
#Adding the new information, if we missed on in our function above than it will be null
for key, currentAssets in newInfo.items():
    sp500.loc[key,"Current Assets"] = currentAssets / 1000000 #We divide by millions so that the number is easier to read
```
