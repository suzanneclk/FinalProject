```python
profitMargin = {}
debtToEquityRatio = {}
returnOnAssets = {}
quickRatio = {}


symbolsNotFoundInLoop = []

""" 
I realized that there are some stock that do not have a balnce sheet I guess for some reason. So I implemented some error handling into our code to handle that - JB

UPDATE
-------

I created a list of the companies that are no longer listed on the S&P 500. This should run without error but if for some reason the lenght of this list is > 0 than we will need to review the list
and make adjustments

"""

for c, r  in sp500.iterrows():
    companyBalanceSheet = yf.Ticker(r[0]).balance_sheet
    companyIncomeStatement = yf.Ticker(r[0]).income_stmt
    
    # print(companyBalanceSheet.index)
    try:
        #This try is becuase there is a small percentage of companies with 0 revenue and negative income, so we will just use a profit margin of 0 becuase they didnt make any revenues
        try:
            pm = companyIncomeStatement.loc["Net Income"][0]/companyIncomeStatement.loc["Total Revenue"][0]
            
            
        except:
            pm = 0
             
        dToE = companyBalanceSheet.loc["Total Liabilities Net Minority Interest"][0]/companyBalanceSheet.loc["Stockholders Equity"][0]
                        
        #Calculating the return on assets for the company
                
        avgAssets = (companyBalanceSheet.loc["Total Assets"][0] + companyBalanceSheet.loc["Total Assets"][1])/2
        roa = companyIncomeStatement.loc["Net Income"][0]/avgAssets
                        
        #Now its time to calculate the quick ratio, this one is a little more simple  
        try: #this try block is here becuase I realized that some companies dont have A.R. becuase thay have a factoring agreement
            qr = (companyBalanceSheet.loc["Cash Cash Equivalents And Short Term Investments"][0] + companyBalanceSheet.loc["Accounts Receivable"][0])/companyBalanceSheet.loc["Current Liabilities"][0]
        
        except:
            qr = (companyBalanceSheet.loc["Cash Cash Equivalents And Short Term Investments"][0])/companyBalanceSheet.loc["Current Liabilities"][0]
            
    except Exception as inst:
        print(r[0])
        print(yf.Ticker(r[0]).income_stmt)
        print("\n", inst.__annotations__)
        break
            
    
    #This is where I add the information we got into the dictionary for use in the function that will follow :)
    profitMargin[r.Symbol] = pm
    debtToEquityRatio[r.Symbol] = dToE
    returnOnAssets[r.Symbol] = roa
    quickRatio[r.Symbol] = qr
        
    
```

```python
# #Adding the new information, if we missed on in our function above than it will be null
def add_info(new_info: dict, columnName: str):
    for key, value in new_info.items():
        sp500.loc[key, columnName] = value
        
add_info(new_info = profitMargin, columnName = "profitMargin")
add_info(new_info = debtToEquityRatio,columnName = "debtToEquityRatio")
add_info(new_info = returnOnAssets, columnName = "returnOnAssets")
add_info(new_info = quickRatio,columnName =  "quickRatio")
        
```
